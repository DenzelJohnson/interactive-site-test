```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import Normalize, BoundaryNorm
from matplotlib.cm import ScalarMappable, get_cmap

# Load your data
data = pd.read_csv('comprehensive_travel_time_adjustments.csv')

# Determine the number of unique origins and destinations
origins = data['Origin'].unique()
destinations = data['Destination'].unique()

# Sort origins and destinations if not already ordered
origins = sorted(origins)
destinations = sorted(destinations)

# Create a large figure to hold all subplots (heatmaps)
fig, axes = plt.subplots(nrows=len(origins), ncols=len(destinations), figsize=(20, 20),
                         sharex='col', sharey='row')  # Share x and y for clean axes management

# Flatten the axes array for easy handling in case of a single row or column
if len(origins) == 1 or len(destinations) == 1:
    axes = axes.flatten()
else:
    axes = axes.ravel()  # Flatten the 2D axes array for easier linear indexing

# Determine the full range of the matrix data
full_range = data.iloc[:, 2:51].values  # Assuming these are the matrix data columns
min_val = np.nanmin(full_range)
max_val = np.nanmax(full_range)

# Define a colormap with 6 discrete colors
n_colors =24
cmap = get_cmap('Spectral', n_colors)  # You can choose any colormap
norm = BoundaryNorm(np.linspace(min_val, max_val, n_colors + 1), n_colors)

# Plotting each 7x7 matrix in the appropriate subplot, use `norm` to normalize colors
for (i, j), ax in zip(np.ndindex(len(origins), len(destinations)), axes):
    sub_data = data[(data['Origin'] == origins[i]) & (data['Destination'] == destinations[j])]
    if not sub_data.empty:
        try:
            matrix_data = sub_data.iloc[0, 2:51].astype(float).values.reshape((7, 7))
            im = ax.imshow(matrix_data, cmap=cmap, norm=norm)
            ax.set_xticks([])  # Hide x-axis ticks
            ax.set_yticks([])  # Hide y-axis ticks
        except ValueError:
            ax.text(0.5, 0.5, 'Data Error', horizontalalignment='center',
                    verticalalignment='center', transform=ax.transAxes)
    else:
        ax.axis('off')

# Setting labels for the first column and top row
for ax, row in zip(axes.reshape(len(origins), -1)[:, 0], origins):
    ax.set_ylabel(row, rotation=90, labelpad=15, verticalalignment='center',fontsize=12, fontweight='bold')

for ax, col in zip(axes.reshape(-1, len(destinations))[0], destinations):
    ax.set_title(col, pad=10,fontsize=12, fontweight='bold')

# Adjust layout and add a colorbar
fig.subplots_adjust(wspace=0.1, hspace=0.1, right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(ScalarMappable(norm=norm, cmap=cmap), cax=cbar_ax)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import Normalize
from matplotlib.cm import get_cmap

# Load your data
data = pd.read_csv('comprehensive_travel_time_adjustments.csv')

# Define the specific origin and destination you are interested in
specific_origin = 'Hamilton'  # Change to your specific origin
specific_destination = 'Chicago'  # Change to your specific destination

# Filter the data for the specific origin-destination pair
sub_data = data[(data['Origin'] == specific_origin) & (data['Destination'] == specific_destination)]


if not sub_data.empty:
    try:
        # Extract the matrix data
        matrix_data = sub_data.iloc[0, 2:51].astype(float).values.reshape((7, 7))

        # Extract column names for uncertainty and coverage
        col_names = data.columns[2:51]  # Adjust index as per your data
        # Parsing the uncertainty and coverage levels from column names
        coverage_levels = [float(name.split('_')[3]) for name in col_names]
        uncertainty_levels = [float(name.split('_')[1]) for name in col_names]

        # Creating unique sorted lists of coverage and uncertainty levels
        unique_coverage_levels = sorted(set(coverage_levels))
        unique_uncertainty_levels = sorted(set(uncertainty_levels), reverse=True)  # Reverse sort for uncertainty

        # Define the colormap
        cmap = get_cmap('viridis')  # Or any other colormap that you prefer
        norm = Normalize(vmin=np.min(matrix_data), vmax=np.max(matrix_data))
        
        # Create the plot
        fig, ax = plt.subplots(figsize=(8, 8))
        cax = ax.imshow(matrix_data, cmap=cmap, norm=norm)

        # Setting the ticks with labels
        ax.set_xticks(np.arange(len(unique_coverage_levels)))
        ax.set_yticks(np.arange(len(unique_uncertainty_levels)))
        ax.set_xticklabels([f"{level:.2f}" for level in unique_coverage_levels])
        ax.set_yticklabels([f"{level:.2f}" for level in unique_uncertainty_levels])

        # Setting axis labels
        ax.set_xlabel('Coverage Level')
        ax.set_ylabel('Uncertainty Level')
        
        # Adding text annotations inside the cells
        for (i, j), val in np.ndenumerate(matrix_data):
            ax.text(j, i, f"{val:.2f}", ha='center', va='center', color='white')
        
        # Colorbar
        fig.colorbar(cax, ax=ax, orientation='vertical')
        plt.show()
        
    
    except ValueError as e:
        print(f"Data Error: {e}")
else:
    print("No data available for this origin-destination pair.")
```
